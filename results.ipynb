```python
import pandas
```

```python
df = pandas.read_csv('Resultats_IA_3.csv')
df.head()
```

```python
matrix =df.pivot_table(index='IA1', columns='IA2', values='%GAGNE')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(matrix)
plt.title("Heatmap des r√©sultats des IA sur un jeu de Hex 3x3")
plt.show()
```
